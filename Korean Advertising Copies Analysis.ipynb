## 2. 카테고리 정리
- [기술] IT + 이동통신 + 휴대폰 + 인터넷 + 통신 + 가전 + 전자 + 자동차
- [취미] 엔터테인먼트 + 도서 + 뮤지컬/연극 + 연극 + 영화 + 스포츠 + 여행
- [패션] 화장품 + 화장 + 미용 + 의류 + 패션
- [건강] 건강 + 의료 + 의약 + 제약
- [공공] 공공 + 공익 + 정부
- [교육] 교육 + 학교
- [금융] 금융 + 은행 + 보험 + 카드
- [정치] 선거 + 정치
- [서비스] 쇼핑 + 홈쇼핑 + 서비스 + 유통
- [생활] 식품 + 음료 + 주류 + 생활
- [건설] 건설 + 인테리어
- [기업] 기업
- [기타] 기타

```python
set(df['category'])
```

```python
tech = ['IT', '이동통신', '휴대폰', '인터넷','통신','가전',' 전자', '자동차']
hobby = ['엔터테인먼트', '도서', '뮤지컬', '연극', '영화', '스포츠', '여행']
fashion = ['화장품', '화장', '미용', '의류', '패션']
health = ['건강', '의료', '의약', '제약']
public = ['공공', '공익', '정부']
edu = ['교육', '학교']
finance = ['금융', '은행', '보험', '카드']
politic = ['선거', '정치']
service = ['서비스', '쇼핑', '홈쇼핑', '유통']
life = ['식품', '음료', '주류', '생활']
build = ['건설', '인테리어']
company = ['기업']
etc = ['기타']
```

```python
def category(cat):
    if any(elem in cat for elem in tech):
        return '기술'
    elif any(elem in cat for elem in hobby):
        return '취미'
    elif any(elem in cat for elem in fashion):
        return '패션'
    elif any(elem in cat for elem in health):
        return '건강'
    elif any(elem in cat for elem in public):
        return '공공'
    elif any(elem in cat for elem in edu):
        return '교육'
    elif any(elem in cat for elem in finance):
        return '금융'
    elif any(elem in cat for elem in politic):
        return '정치'
    elif any(elem in cat for elem in service):
        return '서비스'
    elif any(elem in cat for elem in life):
        return '생활'
    elif any(elem in cat for elem in build):
        return '건설'
    elif any(elem in cat for elem in company):
        return '기업'
    elif any(elem in cat for elem in etc):
        return '기타'
```

```python
df['category'] = df['category'].apply(lambda x : category(x))
```

## 13. 문장 유형
