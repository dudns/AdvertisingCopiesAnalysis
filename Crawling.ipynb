# 광고 카피 크롤링
- 웹사이트 "카피가 강한 세상에"(http://adim21.co.kr) 에 업로드된 광고 슬로건 수집하기

1) 광고 슬로건

2) 해당 슬로건을 이용한 브랜드 

3) 슬로건이 공개된 연도 (광고 언어의 통시적 분석을 위해)

4) 광고 품목 (광고 품목별 비교를 위해)

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import re
```

```python
df = pd.DataFrame(columns=["year", "copy", "brand", "category"])
count = 0 
```

```python
# 1999년 광고부터 수집
# 3918 경계로 형식 달라져서 나눠서 수집 
for num in range(3826,3918):
    url = "http://adim21.co.kr/xe/index.php?mid=head&document_srl=" + str(num)
    with urㄴllib.request.urlopen(url) as url:
        try:
            doc = url.read()
            soup = BeautifulSoup(doc, "html.parser")
            
            year = soup.find_all("h1", class_="np_18px")[0].get_text()
            copies = soup.find_all("div", class_="document_"+str(num)+"_4 xe_content")[0].get_text()
            
            year = re.findall(r'\d{4}년', year)[0]
            copies = copies.split('\n')
            
            for c in copies:
                if c.startswith('*'):
                    category = c.replace('* ', '').replace('\r', '')
                if c.startswith('- '):
                    copy = re.split('-', c, 2)[1]
                    brand = re.split('-', c, 2)[2]

                    df.loc[count] = [year, copy, brand, category]
                    count+=1
            
        except:
            pass 
```

```python
df
```

```python
# 중간 저장 
df.to_csv('copies1.csv', index=False, encoding='utf-8')
```

```python
df2 = pd.DataFrame(columns=["year", "copy", "brand", "category"])
count = 0
```

```python
# 3918부터 달라진 새로운 형식에 맞춰 수집
for num in range(3918, 5000):
    url = "http://adim21.co.kr/xe/index.php?mid=head&document_srl=" + str(num)
    try:
        with urllib.request.urlopen(url) as url:
        
            doc = url.read()
            soup = BeautifulSoup(doc, "html.parser")
            
            year = soup.find_all("h1", class_="np_18px")[0].get_text()
            copies1 = soup.find_all("div", class_="document_"+str(num)+"_4 xe_content")[0].get_text()
            
            year = re.findall(r'\d{4}년', year)[0]
            copies = copies1.split('\n')

            for c in copies:
                if '*' in c:
                    category =c.replace('* ', '')
                if '[' in c:
                    copy = c.split('[')[0]
                    brand = c.split('[')[1].replace(']', '')

                    df2.loc[count] = [year, copy, brand, category]
                    count+=1
            
    except:
        pass 
```

```python
df2.to_csv('copies2.csv', index=False, encoding='utf-8')
```
